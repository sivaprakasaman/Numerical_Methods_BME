```python
#generate LHS samples at lower res for metamodeling

test = lhs(2, samples = 50)
test = bed_min + test*(bed_range);
test = np.asarray(test);

BH_i, BU_i = test.T;
```

```python
#Implement IDW 
# import idw

# idw_tree = idw.tree(np.hstack((BH,BU)),r);

# X2 = np.reshape(test, (2, -1)).T
# print(X2)
# r_i = idw_tree(X2);

# print(r_i)

# Prediction
def distance(x, y, xi, yi):
    d = np.sqrt((x-xi)**2 + (y-yi)**2);
    return(d)

def idwr(x, y, z, xi, yi):
    lstxyzi = []
    for p in range(len(xi)):
        lstdist = []
        for s in range(len(x)):
            d = distance(x[s], y[s], xi[p], yi[p])
            lstdist.append(d)
        sumsup = list((1 / np.power(lstdist, 2)))
        suminf = np.sum(sumsup)
        sumsup = np.sum(np.array(sumsup) * z.T)
        u = sumsup / suminf
        xyzi = u
        lstxyzi.append(xyzi)
    return(lstxyzi)

r_i = idwr(BH,BU,r,BH_i,BU_i);
plot1 = plt.scatter(BH_i,BU_i,c = r_i)
```

```python
#Implement RBF

from scipy.interpolate import Rbf

rbf_i = Rbf(BH, BU, r);

r_i_rbf = rbf_i(BH_i, BU_i);

plot1 = plt.scatter(BH_i,BU_i,c = r_i_rbf)
```
